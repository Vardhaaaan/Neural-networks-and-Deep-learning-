```python
import numpy as np
import tensorflow as tf
X = np.array([[0.3, 0.5, 0.4]], dtype=np.float32)
y = np.array([[1.0]], dtype=np.float32)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(
        2,
        activation='sigmoid',
        input_shape=(3,)
    ),
    tf.keras.layers.Dense(
        1,
        activation='sigmoid'
    )
])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),
    loss='mse'
)
```

```python
history = model.fit(
    X,
    y,
    epochs=1000,
    verbose=0
)
```

```python
output = model.predict(X)
print("Final Output:", output)
```

```python
for i in [0, 100, 300, 500, 700, 900]:
    print(f"Epoch {i}, Loss: {history.history['loss'][i]:.6f}")
```
