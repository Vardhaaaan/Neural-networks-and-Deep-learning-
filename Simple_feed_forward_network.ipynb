```python
import numpy as np

x = np.array([0.6,0.8,0.4])
w=np.array([[0.1,0.3],[0.4,0.5],[0.7,0.6]])
b=np.array([0.03,0.02])
z=np.dot(x,w)+b
print("Output: ",z)
def sigmoid(X):
  return 1/(1+np.exp(-X)) # Changed -x to -X
a=sigmoid(z)
print("Activation Output: ",a)
w2=np.array([[0.4],[0.3]])
b2=np.array([0.06])
z2=np.dot(a,w2)+b2
a2=sigmoid(z2)
print("Final Output: ",a2)
y=1
loss=(y-a2)**2
print("Loss: ",loss)
lr=0.2
dloss_da2=-2*(y-a2)
da2_dz2=a2*(1-a2)
dz2_dw2=a
grad_w2=dloss_da2*da2_dz2*dz2_dw2
w2=w2-lr*grad_w2.reshape(1,2)
print("Updated w2: ",w2)

```
