```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image


# Load pretrained CNN model

model = MobileNetV2(weights="imagenet")


# Load and preprocess image

img_path = "images (2).jpeg"   # put your image here
img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)


# Predict

preds = model.predict(img_array)
decoded = decode_predictions(preds, top=5)[0]


# Determine animal type

dog_classes = [
    "shepherd", "retriever", "terrier", "bulldog",
    "hound", "poodle", "spaniel", "malinois",
    "leonberg", "briard", "chihuahua", "husky"
]

cat_classes = [
    "cat", "persian", "siamese", "tabby", "egyptian"
]

bird_classes = [
    "bird", "parrot", "macaw", "eagle", "sparrow",
    "owl", "cockatoo", "hen", "peacock", "flamingo"
]

animal_type = "Unknown"

for _, name, score in decoded:
    lname = name.lower()

    if any(dog in lname for dog in dog_classes):
        animal_type = "Dog"
        break
    elif any(cat in lname for cat in cat_classes):
        animal_type = "Cat"
        break
    elif any(bird in lname for bird in bird_classes):
        animal_type = "Bird"
        break


# Output

print("Animal Type:", animal_type)
print("\nTop Predictions:")

for _, name, score in decoded:
    print(f"{name} : {score*100:.2f}%")
```
